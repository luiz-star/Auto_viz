#Autoviz

Autoviz é uma biblioteca python de código aberto que trabalha principalmente na visualização da relação dos dados, pode encontrar os recursos mais impactantes e traçar a visualização criativa em apenas uma linha de código. Autoviz é incrivelmente rápido e altamente útil.

```python
#Instalando a Biblioteca
!pip install autoviz
```

```python
#Para usar o autoviz, primeiro precisamos importar a classe autoviz e instanciá-la.
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

```python
#Após iniciar a classe Autoviz,
# precisamos apenas executar um comando que irá criar uma visualização do conjunto de dados.
df = AV.AutoViz ('https://raw.githubusercontent.com/Gabrielmarcial/Data-Marte/main/datasets/train.csv')
```
